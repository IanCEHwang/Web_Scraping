### 1.please get Selenium to work on your system. e., try to code something up in Java or Python that starts a browser of your choice, navigates to google.com, and searches for "askew" as well as "google in 1998" (separate searches!)

```python
### URL to google
url = 'https://www.google.com'

### directory to my chrome driver
PROJECT_ROOT = os.path.abspath(os.path.dirname('./Users/huangjien/Desktop/chromedriver_mac64'))
DRIVER_BIN = os.path.join(PROJECT_ROOT, "chromedriver")

### execute chrome driver
driver = webdriver.Chrome(executable_path = DRIVER_BIN)

### go to Google
driver.get(url)
```

```python
### connect input box
input_box = driver.find_element(By.CSS_SELECTOR , "input[type=text]")

### send search terms 
### askew
input_box.send_keys("askew")

### send search
input_box.submit()

### sleep for 1 second
time.sleep(1)
```

```python
### connect input box
input_box = driver.find_element(By.CSS_SELECTOR , "input[type=text]")

### clear previous search
input_box.clear()

### send search terms 
### google in 1998
input_box.send_keys("google in 1998")

### send search
input_box.submit()

### stay for 5 seconds
time.sleep(5)

### close chrome
driver.close()
```

### 2. write a script that goes to bestbuy.com, clicks on Deal of the Day, reads how much time is left for the Deal of the Day and prints the remaining time to screen (console), clicks on the Deal of the Day (the actual product), clicks on its reviews, and saves the resulting HTML to your local hard drive as "bestbuy_deal_of_the_day.htm"


```python
### URL to bestbuy
url = 'https://www.bestbuy.com'

### directory to my chrome driver
PROJECT_ROOT = os.path.abspath(os.path.dirname('./Users/huangjien/Desktop/chromedriver_mac64'))
DRIVER_BIN = os.path.join(PROJECT_ROOT, "chromedriver")

### execute chrome driver
driver = webdriver.Chrome(executable_path = DRIVER_BIN)

### go to bestbuy
driver.get(url)

### sleep for 1 second
time.sleep(1)
```

```python
### connect button "Deal of the Day"
Deal_of_the_D = driver.find_element(By.XPATH , "//a[text() = 'Deal of the Day' ]")

### click on button
Deal_of_the_D.click()

### sleep for 1 second
time.sleep(1)
```

```python
### get the page source code
html = driver.page_source

### parse source code to html format
bs4_html = BeautifulSoup(html , 'html.parser')

### find clock element
clock = bs4_html.select("div[class = 'countdown-clock']")

### print remaining time
print("Remaing time :")
print(clock[0].text)

### sleep for 1 second
time.sleep(1)
```

```python
### connect button "Deal of the Day"
Deal_of_the_D = driver.find_element(By.CSS_SELECTOR , "a[class = 'wf-offer-link v-line-clamp ' ]")

### click on button
Deal_of_the_D.click()

### sleep for 1 second
time.sleep(1)
```

```python
### connect button "Review"
review = driver.find_element(By.XPATH , "//div[@class='user-generated-content-ratings-and-reviews']")

### click on review
review.click()

### sleep
time.sleep(1)

### click on review again to actually expand
review.click()
```
